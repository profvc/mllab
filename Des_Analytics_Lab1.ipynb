#**Descriptive Analytics**

Descriptive analytics is the process of summarising data to gain insights into the data set. It is used to uncover patterns and trends in data, enabling organisations to make data-driven decisions. Descriptive analytics can be used to identify customer preferences, improve customer experience, and identify potential areas of improvement or opportunity.

**Import/install necessary packages**

```python
 import pandas as pd
```

```python
data = {'BL': [5.6,5,5.8,3,7,4.6,7,5.6,5,5.8], 
        'BAS': [6,7,5.25,5.5,6.5,4.75,6,5.5,6.25,7],
        'BAW': [4,5,6,2,7,4.25,6,5,5.25,6],
        'PQ': [6,7,6,5.25,5.25,6,1,5.25,6,6.25],
        'OBE': [6,5,6,5.5,7,6,5,6,7,6.5]}
df = pd.DataFrame(data)
```

**Skewness and Kurtosis**

Skewness is a measure of the asymmetry of a data distribution, while kurtosis is a measure of the peakedness of a data distribution.
Both measures can help identify underlying patterns in data that can indicate potential outliers or errors in data collection.


```python
df.skew()
```

```python
df.kurtosis()
```

# **Chart visualisation**

**Display Histogram - Overall Brand Equity**

A histogram is a graphical representation of data that is used to show the frequency distribution of a given data set.
It is a bar chart that groups a range of data into different bins and displays the number of data points that fall within each bin.


```python
df.OBE.plot.hist()
```

**Display a Density Plot**

KDE Density Plot: 
A KDE density plot is a graphical representation of data that displays the probability density of a given data set.
It is a smooth curve that shows the estimated probability density of each data point.


```python
df.OBE.plot.kde()
```

```python
df.OBE.plot.kde(ind=[1, 2, 3, 4, 5, 6, 7])
```

**Display a Line Chart - OBE**

A line plot is a graph that shows a set of data points connected by straight lines. It is used to observe trends and patterns in the data, and can be used to compare multiple data sets on the same graph.


```python
df.OBE.plot()
```

**Display a Box Plot - OBE**

A box plot is a graphical representation of data that shows the median, quartiles, maximum, and minimum values in a dataset.
It is used to compare multiple datasets, identify outliers, and assess overall data distribution and spread.


```python
df.plot.box()
```

```python
df.plot.box(vert=False);
```

```python
df.OBE.plot.box()
```

**Display an Area Plot**

An area plot is a two-dimensional graph that displays quantitative data through a filled polygon, which is created by connecting a sequence of points.
It is used to visualise the changes in data over a certain period of time, track cumulative totals, and compare the relationship between two or more variables.


```python
df.plot.area()
```

**Display a Pie Plot**

Pie plots are two-dimensional graphs that display data as a circle divided into sections, with each section representing a proportion of the whole.
Pie plots are used to compare the relative size of different categories or groups within a dataset.


```python
series = df.sum()
series.plot.pie(figsize=(4, 4));
```

**Scatter plot using pairs of variables (Brand Association vs Overall Brand Equity)**

Scatter plots are two-dimensional graphs that display data as a collection of points, with each point representing the values of two variables.
The position of the point on the graph indicates how the two variables are related, allowing for easy identification of trends and patterns.


```python
df.plot.scatter(x='BAS',y='OBE')
```

```python
df.plot(x='BAS',y='OBE',kind="scatter")
```

# **Conclusion**

Descriptive analytics is a powerful tool for uncovering patterns and trends in data. It can be used to identify customer preferences, improve customer experience, and identify potential areas of improvement or opportunity. However, it is limited in its ability to predict future outcomes or behaviour, and identify causal relationships between variables.
